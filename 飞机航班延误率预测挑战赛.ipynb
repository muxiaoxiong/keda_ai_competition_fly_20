```python
import pandas as pd 
pd.set_option('display.max_columns', None)
import numpy as np
import os 

from sklearn.model_selection import train_test_split
```

# 查看数据

```python
train=pd.read_csv('./data/train.csv')
test=pd.read_csv('./data/test.csv')
train=train.drop('ID',axis=1)
test=test.drop('ID',axis=1)
test
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
label = 'LABEL'
save_path = './model'
metric = 'roc_auc'
predictor = TabularPredictor(label=label, path=save_path,eval_metric=metric).fit(train,presets='good_quality')
predictor.leaderboard(test, silent=True)
# y_pred = predictor.predict_proba(test)
# perf = predictor.evaluate_predictions(y_true=test, y_pred=y_pred, auxiliary_metrics=True)
```

```python
y_pred = predictor.predict_proba(test)
```
